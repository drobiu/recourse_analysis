```python
import matplotlib.pyplot as plt
import json
```

```python
f = open('../results/261733.json')
```

```python
j_obj = json.load(f)
```

```python
j_obj
```

```python
methods = ['CLUE', 'Wachter']
```

```python
for method in methods:
    col = '#c78f1e' if method == 'Wachter' else '#0096f0'
    plt.plot(j_obj[method]['disagreement'][:50], c=col, linewidth=2)
plt.xlabel('Iteration')
plt.ylabel('Disagreement')
plt.savefig('disagreement.png')
plt.show()
```

```python
for method in methods:
    col = '#c78f1e' if method == 'Wachter' else '#0096f0'
    plt.plot(j_obj[method]['mmd'][:50], c=col, linewidth=2)
plt.xlabel('Iteration')
plt.ylabel('MMD')
plt.savefig('mmd.png')
plt.show()
```

```python
fig = plt.figure(constrained_layout=True, figsize=(7, 6))
axs = fig.subplot_mosaic([['top_left', 'top_right'],['bottom_left', 'bottom_right']],
                          gridspec_kw={'width_ratios':[1, 1]})

it = 50
for method in methods:
    col = '#c78f1e' if method == 'Wachter' else '#0096f0'
    axs['top_left'].plot(j_obj[method]['mmd'][:it], c=col, linewidth=2)
    axs['top_right'].plot(j_obj[method]['disagreement'][:it], c=col, linewidth=2)
    axs['bottom_left'].plot(j_obj[method]['model_mmd'][:it], c=col, linewidth=2)
    axs['bottom_right'].plot(j_obj[method]['prob_mmd'][:it], c=col, linewidth=2)
axs['top_left'].set_title('MMD')
axs['top_right'].set_title('Disagreement')
axs['bottom_left'].set_title('Boundary MMD')
axs['bottom_right'].set_title('Gradient MMD')
```
