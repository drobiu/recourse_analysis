```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
np.random.seed(2)

x, y = make_moons(200, noise=0.15)

plt.scatter(x[:, 0], x[:, 1], c=y)
plt.show()

x = np.around(x, decimals=4)

with open('datasets/moons.csv', 'w+', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['feature1', 'feature2', 'target'])
    
    for i in range(len(x)):
        writer.writerow([*x[i], y[i]])
        
    f.close()
```
